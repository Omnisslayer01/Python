```python
#chatgpt solution for table from 1 to 10
def display_tables():
    # Loop through numbers 1 to 10 to create each table
    for i in range(1, 11):
        print(f"Multiplication Table for {i}:")
        # Loop through numbers 1 to 10 to create the rows of the table
        for j in range(1, 11):
            print(f"{i} x {j} = {i * j}")
        print("-" * 20)  # Separator line for readability

# Call the function to display the tables
display_tables()
```

```python
#table from 1 to 10
def multiplier(i):
    for i in range(1,11):
        print("-"*20)
        print(f"The table of {i} is:")
        for j in range(1,11):
            print(f"{i} x {j} = {i}*{j}")
            
multiplier(i)
```
